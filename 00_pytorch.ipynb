### Tensor Datatypes

```python
# float tensor
from torch import float16
float_tensor = torch.tensor([3.0, 4.0, 5.0], 
                            dtype=torch.float32,
                            device=None,
                            requires_grad=False)
float_tensor.dtype
```

```python
float_half_tensor = torch.tensor([3.0, 4.5, 5.67], 
                                dtype=torch.half)
# float16 tensor can also be created using torch.half
float_half_tensor.dtype
```

```python
(float_tensor * float_half_tensor).dtype
# by default, PyTorch will upcast to the higher precision 
# data type when performing operations on tensors 
# of different data types. In this case, the result of 
# multiplying a float32 tensor with a float16 tensor 
# will be a float32 tensor.
```

```python
# integer tensor
int_tensor = torch.tensor([3, 4, 5],
                          dtype=torch.int32)
int_tensor, int_tensor.dtype

# Example of exceeding limit of int16
big_int_tensor = torch.tensor([30000, 40000, 50000], dtype=torch.int16)
print(big_int_tensor) # this throws an error because the values exceed the limit of int16
```

```python
print(int_tensor * float_half_tensor)
(int_tensor * float_half_tensor).dtype
# when multiplying an integer tensor with a float tensor,
# the result will be a float tensor, as PyTorch will upcast
# the integer tensor to the higher precision float tensor.
```

### Tensor Operations
1. Addition
2. Subtraction
3. Element-wise Multiplication
4. Division
5. Matrix Multiplication

```python
# Create a tensor
tensor = torch.tensor([1, 2, 3])
# Adding a scalar to a vector
tensor + 10 
# This adds 10 to each element in the tensor, 
# demonstrating broadcasting.
```

```python
# Multiplying a tensor by a scalar
tensor * 10
```

```python
tensor
# The original tensor remains unchanged because 
# PyTorch operations return new tensors by default,
# rather than modifying the original tensor in place.
```

```python
tensor = tensor * 10
tensor 
# the original tensor is now updated to the new 
# value after multiplication
```

```python
tensor / 10
```

```python
# PyTorch in-built functions for some operations
new_tensor = torch.tensor([1, 2, 3, 4])
print(torch.add(new_tensor, 10)) # adds 10 to each element in new_tensor
print(torch.sub(new_tensor, 10)) # subtracts 10 from each element in new_tensor
print(torch.mul(new_tensor, 10)) # multiplies each element in new_tensor by 10
print(torch.div(new_tensor, 10)) # divides each element in new_tensor by 10
print(torch.pow(new_tensor, 2)) # raises each element in new_tensor to the power of 2
```

### Matrix Multiplication
#### Two Main Ways :
1. Element-wise multiplication
2. Matrix Multiplication (Most common in neural networks)

```python
# Element-wise multiplication
print(f'{tensor} * {tensor} = {tensor * tensor}')
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
# This performs matrix multiplication (also known as dot product) 
# between the tensor and itself
# tensor = [10, 20, 30] 
# dot product = (10*10) + (20*20) + (30*30) = 100 + 400 + 900 = 1400
```

```python
# Matrix multiplication can also be done by using @ operator
tensor @ tensor
```

```python
# Shape compatibility issues in matrix multiplication
tensor_a = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]])

tensor_b = torch.tensor([[7, 8],
                         [9, 10], 
                         [11, 12]])

print(torch.matmul(tensor_a, tensor_b))
# This will throw an error because the inner dimensions of 
# the tensors do not match for matrix multiplication.
```

```python
# We can solve this by transposing one of the tensors
# Transposing tensor_b 
tensor_b.T
print(f'Original tensor B shape   = {tensor_b.shape}')
print(f'Transposed tensor B shape = {tensor_b.T.shape}')
print(f"Tensor A shape            = {tensor_a.shape}")
```

```python
# Now multiplication will work
torch.matmul(tensor_a, tensor_b.T)
```

### Reshaping, stacking, suqeezing and unsqueezing tensors

```python
x = torch.arange(1, 10)
x, x.shape
```

```python
reshaped = x.reshape(1,9)
reshaped, reshaped.shape
```

```python
z = x.view(1, 9)
z.shape
```

```python
z[:, 0] = 5
z, x
```

```python
# Stacking tensors
stacked = torch.stack([x, x, x], dim=0)
stacked
```

```python
# squeezing tensors
squeezed = torch.squeeze(reshaped)
# squeezing removes dimensions of size 1 from the tensor, 
# so the shape of the squeezed tensor will be (9,) instead of (1, 9)
print(f"Original Shape & Size : {reshaped}, {reshaped.shape}\nSqueezed Shape & Size : {squeezed}, {squeezed.shape}")
```

```python
# Unsqueezing tensors
unsqueezed = torch.unsqueeze(squeezed, dim=0)
print(f"Original Shape & Size   : {squeezed}, {squeezed.shape}\nUnsqueezed Shape & Size : {unsqueezed}, {unsqueezed.shape}")
```

```python
original = torch.rand(size=(224, 224, 3))
print("Original Shape : ", original.shape)
print("First element before permuting : ", original[0, 0, 0])

# Permuting tensors
permuted = original.permute(2, 0, 1)
print("First element after permuting : ", permuted[0, 0, 0])
print("Permuted Shape : ", permuted.shape)
```
