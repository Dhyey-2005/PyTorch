### Random Tensors

```python
# Create a random tensor of shape (3, 4)
random_tensor = torch.rand(2, 4, 4)
random_tensor
```

```python
random_tensor.ndim
```

```python
# Create a random tensor with similar shape to image tensor
random_image_tensor = torch.rand(size=(224, 224, 3)) # heght, width, color channels (R, G, B)
print(random_image_tensor.shape, random_image_tensor.ndim)
```

### Zeros and Ones

```python
zeros = torch.zeros(3, 4)
zeros
```

```python
ones = torch.ones(3, 4)
ones
```

### Creating a range of tensors and tensors-like

```python
# using torch.arange()
arr = torch.arange(start=-20, end=50, step=7)
arr
```

```python
# Creating tensors like
tensor_like = torch.zeros_like(input=arr)
tensor_like
```
