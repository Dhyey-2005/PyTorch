### Tensor Datatypes

```python
# float tensor
from torch import float16
float_tensor = torch.tensor([3.0, 4.0, 5.0], 
                            dtype=torch.float32,
                            device=None,
                            requires_grad=False)
float_tensor.dtype
```

```python
float_half_tensor = torch.tensor([3.0, 4.5, 5.67], 
                                dtype=torch.half)
# float16 tensor can also be created using torch.half
float_half_tensor.dtype
```

```python
(float_tensor * float_half_tensor).dtype
# by default, PyTorch will upcast to the higher precision 
# data type when performing operations on tensors 
# of different data types. In this case, the result of 
# multiplying a float32 tensor with a float16 tensor 
# will be a float32 tensor.
```

```python
# integer tensor
int_tensor = torch.tensor([3, 4, 5],
                          dtype=torch.int32)
int_tensor, int_tensor.dtype

# Example of exceeding limit of int16
big_int_tensor = torch.tensor([30000, 40000, 50000], dtype=torch.int16)
print(big_int_tensor) # this throws an error because the values exceed the limit of int16
```

```python
print(int_tensor * float_half_tensor)
(int_tensor * float_half_tensor).dtype
# when multiplying an integer tensor with a float tensor,
# the result will be a float tensor, as PyTorch will upcast
# the integer tensor to the higher precision float tensor.
```
